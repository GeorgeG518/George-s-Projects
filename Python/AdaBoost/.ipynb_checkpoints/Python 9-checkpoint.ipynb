### Import Python Packages

```python
# Code Block 0
import numpy as np 
import pandas as pd 
import seaborn as sns

from matplotlib import pyplot as plt
from matplotlib import style

from sklearn import tree
from sklearn.tree import plot_tree, DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

<a id="1"></a>
### Set Display
<a href = "#99">Top</a>

```python
# Code Block 1
# Set display to show up to 100 columns in dataframe
pd.set_option('display.max_columns',100) 

# Set plot style
plt.style.use('seaborn')
```

<a id="2"></a>
### Load the Data and Evaluate Fields
<a href = "#99">Top</a>

```python
# Code Block 2
# Read CSV file into dataframe
df_cellco = pd.read_csv("Cellco Churn Data - Features.csv",encoding="latin1")
```

```python
# Code Block 3
# Fields and data that are available in the dataframe ... and position
df_cellco.info()
```

```python
# Code Block 4
# What does the actual data look like?
df_cellco.sample(10)
```

```python
# Code Block 5
# Field names for access
df_cellco.columns
```

<a id="3"></a>
### Determine Baseline Accuracy
<a href = "#99">Top</a>

```python
# Code Block 6
# Baseline accuracy is if we predicted all individuals would remain or churn
# This is the standard that needs to be improved upon
df_cellco['Retained'].value_counts(normalize=True).mul(100).round(1).astype(str) + '%'
```

<a id="4"></a>
### Create Training (Validation) and Testing Datasets
<a href = "#99">Top</a>

```python
# Code Block 7
# Create a list of features to be utilized
X_names = df_cellco.columns.values.tolist()
X_names.remove('Customer ID')
X_names.remove('Retained')

# Create X matrix of features
X = df_cellco[X_names].copy()

# Create y vector of retained results
y = df_cellco[['Retained']].copy()
```

```python
# Code Block 8
# Split the data into 80% training and 20% testing data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .20, random_state=42)
```

<a id="5"></a>
### Create and Evaluate a Simple Defined AdaBoost Model
<a href = "#99">Top</a>

```python
# Code Block 9
# Define an AdaBoost model with 5 stumps and a learning rate of 0.5
weights = {0:.75, 1:.25}
adaBoost = AdaBoostClassifier(DecisionTreeClassifier(max_depth=1, class_weight=weights), 
                              n_estimators=5, learning_rate=0.5, random_state=42)

# Fit AdaBoost model
ab_simple = adaBoost.fit(X_train,y_train.values.ravel())

# Show overall accuracy for AdaBoost model
print('Overall Accuracy:  '+'{:.2%}'.format(ab_simple.score(X_train,y_train)))
```

```python
# Code Block 10// good place to look at to finish this assignment
# Show the five stumps in the AdaBoost model
fig, axes = plt.subplots(nrows = 1,ncols = 5,figsize = (27,4), dpi=400)
 
for i in range(0,5):
    tree.plot_tree(ab_simple.estimators_[i], feature_names=X_names, class_names=['Churned', 'Retained'], 
                   rounded=True, precision=3, filled=True, fontsize=12, ax = axes[i])
    axes[i].set_title('Stump: ' + str(i), fontsize = 20)

# Save file with rendering of trees
plt.savefig('ab_simple_5stumps.jpg')
```

##### Evaluate Performance of Simple AdaBoost Model

```python
# Code Block 11
# Predict outcomes using AdaBoost on test data
y_pred_test = ab_simple.predict(X_test)

# Create a confusion matrix showing the predicted and true results
cm = confusion_matrix(y_test, y_pred_test)

# Create a heat map showing the results of the confusion matrix for the test data
fig, axes = plt.subplots(nrows=1, ncols=1, sharex=True, figsize=(5, 3))
fig.suptitle('Confusion Matrix of Cellco Attrition\nUsing 5-Stump AdaBoost Model on Test Data', 
             fontsize=16, y=1.05)
ax = plt.subplot()
sns.heatmap(cm, annot=True, cmap="Blues", annot_kws={"size": 16}, ax=ax, fmt="g")
ax.set_xlabel('Predicted', fontsize=15)
ax.set_ylabel('True', fontsize=15)
ax.xaxis.set_ticklabels(['Churned', 'Retained'], fontsize=12)
ax.yaxis.set_ticklabels(['Churned', 'Retained'], fontsize=12, va='center')
plt.show()
```

```python
# Code Block 12
# Show statistics on AdaBoost model performance on test data
print("Results from 5-Stump AdaBoost Model - Testing Data")
print("Accuracy: "+"{:.2%}".format(accuracy_score(y_test, y_pred_test)))
print("Report: \n", classification_report(y_test, y_pred_test, 
                                          target_names=['Churned', 'Retained'], digits=3)) 

# Show results from training data
y_train_pred = ab_simple.predict(X_train) 
print("\nResults from 5-Stump AdaBoost Model - Training Data")
print("Accuracy: "+"{:.2%}".format(accuracy_score(y_train, y_train_pred)))
```

<a id="6"></a>
### Finding the Best Hyperparameters for AdaBoost
<a href = "#99">Top</a>

```python
# Code Block 13
# Define model settings for search
model = AdaBoostClassifier(DecisionTreeClassifier(max_depth=1, class_weight=weights), random_state=42)

# Set up Search Grid possibilities
search_grid = {'n_estimators': [int(x) for x in np.linspace(start = 8, stop = 200, num = 25)],
               'learning_rate': [round(x,1) for x in np.linspace(start = 0.1, stop = 2, num = 20)]}
```

```python
# Code Block 14
# Random search of parameters, using 3 fold cross validation, 
# Search across 300 combinations of hyperparameters for best accuracy.
# Use all but one CPU core to do this
ab_grid = RandomizedSearchCV(estimator=model, param_distributions=search_grid, 
                             cv=3, n_iter=300, n_jobs=-2, scoring='accuracy')

# Fit the grid search model
ab_grid.fit(X_train, y_train.values.ravel())

# Show hyperparameters from best AdaBoost model 
ab_grid.best_params_
```

##### Evaluate Performance of Best AdaBoost Model

```python
# Code Block 15
# Define an AdaBoost model using "best" hyperparameter settings from grid search
adaBoost = AdaBoostClassifier(DecisionTreeClassifier(max_depth=1, class_weight=weights), 
                              n_estimators=24, learning_rate=1.3, random_state=42)

# Fit "best" AdaBoost model
ab_best = adaBoost.fit(X_train,y_train.values.ravel())

# Predict outcomes using AdaBoost on test data
y_pred_test = ab_best.predict(X_test)

# Create a confusion matrix showing the predicted and true results
cm = confusion_matrix(y_test, y_pred_test)

# Create a heat map showing the results of the confusion matrix for the test data
fig, axes = plt.subplots(nrows=1, ncols=1, sharex=True, figsize=(5, 3))
fig.suptitle('Confusion Matrix of Cellco Attrition\nUsing "Best" AdaBoost Model on Test Data', 
             fontsize=17, y=1.05)
ax = plt.subplot()
sns.heatmap(cm, annot=True, cmap="Blues", annot_kws={"size": 16}, ax=ax, fmt="g")
ax.set_xlabel('Predicted', fontsize=15)
ax.set_ylabel('True', fontsize=15)
ax.xaxis.set_ticklabels(['Churned', 'Retained'], fontsize=12)
ax.yaxis.set_ticklabels(['Churned', 'Retained'], fontsize=12, va='center')
plt.show()
```

```python
# Code Block 16
# Show statistics on "best" AdaBoost model performance on test data
print('Results from "Best" AdaBoost Model - Testing Data')
print('Accuracy: '+'{:.2%}'.format(accuracy_score(y_test, y_pred_test)))
print('Report: \n', classification_report(y_test, y_pred_test, 
                                          target_names=['Churned', 'Retained'], digits=3)) 

# Show results from training data
y_train_pred = ab_best.predict(X_train) 
print('\nResults from "Best" AdaBoost Model - Training Data')
print('Accuracy: '+'{:.2%}'.format(accuracy_score(y_train, y_train_pred)))
```

```python
# Code Block 17
# Show overall importance of the features in the "best" AdaBoost model
feature_imp = pd.Series(ab_best.feature_importances_, index=X_names).sort_values(ascending=False)

# Create a bar plot of feature importance
fig, axes = plt.subplots(sharex=True, figsize=(8, 10))
sns.barplot(x=feature_imp, y=feature_imp.index, palette="Blues_r")
plt.ylabel('Features', fontsize = 14, fontweight='bold')
plt.xlabel('Feature Importance Score', fontsize=14, fontweight='bold')
plt.xticks(fontsize = 13, ha = "center")
plt.suptitle("AdaBoost Feature Importance", fontweight='bold', fontsize=20, y=.92)
plt.show()
```

```python
# Code Block 19
# Predict probabilty that customer will churn from "best" AdaBoost model
newObs_PredProb = ab_best.predict_proba(df_predNew)[:,0][0]
print("Probabilty that Customer will Churn: "+"{:.2%}".format(newObs_PredProb))
```
