##### Scatterplot of Number Accommodated and Beds, with additional data cleansing

```python
# Create regression plot of Number of Beds and Number Accommodated
plt.figure(figsize=(16,8)) 

chart = sns.regplot(x='Beds', y='Accommodates', data=df_airbnb, ci=None, 
            scatter_kws = {"color":"b","alpha":0.25,"s":100})

chart.grid(False)

plt.suptitle('Boston Airbnb Rentals', fontweight='bold', 
          fontsize='20', horizontalalignment='center', y=.96)

plt.title('Scatterplot of Number of Beds and Number Accommodated with Line', fontweight='bold', 
          fontsize='14', ha='center', y=1.01)

plt.xlabel('Number of Beds', fontweight='bold', 
           fontsize='14', ha='center')

plt.ylabel('Number Accommodated', fontweight='bold', 
           fontsize='14', ha='center')
plt.show()

# This chart seems to indicate additional data cleansing might be necessary on number of beds
```

```python
# Calculate and show Pearson's and Spearman's correlation between Number of Beds and Number Accommodated
print('Correlation Coefficients between between the Number of Beds and Number Accommodated')
pearsoncorr = df_airbnb['Beds'].corr(df_airbnb['Accommodates'])
print('Pearson correlation: %.3f' % pearsoncorr)
spearmancorr, spear_p = spearmanr(df_airbnb['Beds'], df_airbnb['Accommodates'])
print('Spearmans correlation: %.3f' % spearmancorr)
```

### Create Initial Linear Regression Model
<a href = "#99">Top</a>

```python
# Drop rows with missing data ... regression requires all data be complete
df_airbnb.dropna(inplace = True)
```

```python
# Set up dataframe with target
y = df_airbnb[['Price']].copy()

# Set up dataframe with features
X = df_airbnb[['Host Response Rate', 'Host Acceptance Rate', 'Host Listings Count', 'Accommodates', 
               'Bathrooms', 'Bedrooms', 'Beds', 'Security Deposit', 'Cleaning Fee', 'Guests Included',
               'Extra People', 'Minimum Nights','Number Of Reviews', 'Reviews Per Month', 'superhost_Yes',
               'verified_Yes', 'property_Bed & Breakfast', 'property_Boat', 'property_Condominium',
               'property_House', 'property_Loft', 'property_Townhouse', 'property_Villa',
               'room_Private room', 'room_Shared room', 'bed_Airbed', 'bed_Couch', 'bed_Futon',
               'bed_Pull-out Sofa', 'instabook_Yes', 'cancel_flexible', 'cancel_moderate',
               'cancel_super_strict_30', 'citybins_Brookline', 'citybins_Cambridge', 'citybins_Charlestown',
               'citybins_Jamaica Plain', 'citybins_Other', 'Review Scores Rating', 'Review Scores Accuracy', 
               'Review Scores Cleanliness', 'Review Scores Checkin', 'Review Scores Communication',
               'Review Scores Location', 'Review Scores Value', 'Guests per Bed', 'Accommodates per Bathroom']].copy()

# Guests Not Included was removed due to linear dependency (related to Guests Included and Accommodates)
# Base levels for one-hot encoded features were removed (property_Apartment, room_Entire home/apt, bed_Real Bed,
#   cancel_strict, citybins_Boston, 
```

```python
# Add a constant to the data so that a regression line intercept can be calculated
X = X.assign(constant=1)

#Make a list of all of the columns in the df
cols = list(X.columns.values) 

# Remove constant from the end of the list
cols.pop(cols.index('constant')) 

#Create new dataframe with columns in the order you want ... constant first
X = X[['constant'] + cols] 
```

```python
# Create regression line on regression data
reg = sm.OLS(y, X).fit()

# Create predictions in a list
pred = reg.predict(X) 

# Create residuals in a list
resid = reg.resid

# Show summary statistics for regression
reg.summary()
```

<a id="6"></a>
### Evaluate Regression Results and Pare Model
<a href = "#99">Top</a>

```python
# Evaluate Variance Inflation Factor for X features
vif = pd.DataFrame()
vif["Feature"] = X.columns
vif["VIF Factor"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

# Show features with VIF above 5
index_names = vif[vif['VIF Factor'] < 5 ].index
vif.drop(index_names, inplace=True)
round(vif, 1).sort_values(by = 'VIF Factor', ascending = False)
```

```python
# Drop likely offending variable based on VIF analysis from datasets
X = X.drop(['Accommodates per Bathroom'], axis = 1)
```

```python
# Create regression line on regression data
reg = sm.OLS(y, X).fit()

# Create predictions in a list
pred = reg.predict(X) 

# Create residuals in a list
resid = reg.resid

# Show summary statistics for regression
reg.summary()
```

```python
# Evaluate Variance Inflation Factor for X features
vif = pd.DataFrame()
vif["Feature"] = X.columns
vif["VIF Factor"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

# Show features with VIF above 5
index_names = vif[vif['VIF Factor'] < 5 ].index
vif.drop(index_names, inplace=True)
round(vif, 1).sort_values(by = 'VIF Factor', ascending = False)
```

```python
# Create plot of residuals along with distributions
plt.figure(figsize=(16,8))
sns.jointplot(x=pred, y=resid, kind = "reg", ratio=5, height = 8, 
              line_kws={'color': 'black'})
plt.xlabel('Predicted Daily Rental Price', fontweight='bold', fontsize='14', horizontalalignment='center')
plt.ylabel('Residual', fontweight='bold', fontsize='14', horizontalalignment='center')
plt.suptitle('Residual Plot with Distributions       ',  fontweight='bold', 
          fontsize='18', horizontalalignment='center', y = 1.02)
plt.show()
```

```python
# Create a final set of features based on backwards stepwise removal of insignificant features
X_stepback = X[['constant', 'Host Response Rate', 'Accommodates', 'Bathrooms', 'Bedrooms',
                'Beds', 'Security Deposit', 'Cleaning Fee', 'Guests Included',
                'Minimum Nights', 'Reviews Per Month', 'superhost_Yes', 'property_House', 
                'room_Private room', 'room_Shared room', 'bed_Airbed', 'bed_Futon',
                'instabook_Yes', 'cancel_moderate', 'cancel_super_strict_30', 
                'citybins_Jamaica Plain', 'citybins_Other', 'Review Scores Rating', 
                'Review Scores Cleanliness','Review Scores Location', 'Review Scores Value',
                'Guests per Bed']].copy()

y_stepback = y.copy()
```

```python
# Create regression line on stepwise data
reg = sm.OLS(y_stepback, X_stepback).fit()

# Create predictions in a list
pred = reg.predict(X_stepback) 

# Create residuals in a list
resid = reg.resid

# Show summary statistics for regression
reg.summary()
```

### Use Regression Results to Predict for New Listing

```python
# Create empty copy of dataframe used in model
df_predNew = X_stepback.head(0).copy()

# Show column names to help input new observation for prediction
df_predNew.columns.values
```

```python
# Create observation for the following prediction:
# Property has a response rate of .8, accommodates 4, 2 bathrooms, 2 bedrooms, 2 beds, $450 security deposit,
#   $120 cleaning fee, 4 guests included, 3 nights minimum, not superhosted, is a loft with the entire place rented,
#   real beds, allows instant booking, has a strict cancellation policy, in Boston, has 9.8 on a rescaled review rating,
#   9.7 on cleanliness rating, 9 on location rating, 8 on values rating, and has 2 guests per bed
# First field, constant, should have a value of 1
df_predNew.loc[0] = [1, .8, 4, 2, 2, 2, 450, 120, 4, 3, 2, 0, 0, 1, 0, 0, 0, 1, 0, 0, 0, 0, 9.8, 9.7, 9, 8, 2]
df_predNew
```

```python
# Create and show prediction results for new observation
prediction = reg.get_prediction(df_predNew)
predResults = prediction.summary_frame(alpha=.05)
predResults
```
